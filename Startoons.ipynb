```python
import numpy as np
import matplotlib.pyplot as plt

# Load the data from the text file
data = np.loadtxt('/content/drive/MyDrive/startoons/Data_1.txt')

# Find the maxima and minima peaks
peaks = []
mins = []
for i in range(1, len(data) - 1):
    if data[i-1] < data[i] > data[i+1]:
        peaks.append(i)
    elif data[i-1] > data[i] < data[i+1]:
        mins.append(i)

# Plot the data with maxima and minima peaks
plt.figure(figsize=(10,6))
plt.plot(data, label='Signal')
plt.plot(peaks, data[peaks], 'ro', label='Maxima Peaks')
plt.plot(mins, data[mins], 'bo', label='Minima Peaks')
plt.xlabel('Index')
plt.ylabel('Signal Value')
plt.title('Maxima and Minima Peaks of the Signal')
plt.legend()
plt.grid(True)
plt.show()
```
